```python
from bs4 import BeautifulSoup

# 读取 HTML 文件的路径
file_path = r"E:\JHU\SoftwareCarpentry\Final_Project\2023 Formula 1 World Championship Race Schedule on ESPN Platforms - ESPN Press Room U.S..html"

# Open and read the HTML file
with open(file_path, 'r', encoding='utf-8') as file:
    contents = file.read()
# Create a BeautifulSoup object and specify the parser
soup = BeautifulSoup(contents, 'html.parser')

# Extract the race schedules
race_schedule = []
table_rows = soup.find_all('tr')
for row in table_rows:
    # Extract all cells in the row
    cells = row.find_all('td')
    # Check if the row contains the expected number of cells
    if len(cells) > 2:  # Adjust the number based on the actual content
        # Extract the text from the cells
        session = cells[0].get_text(strip=True)
        date = cells[1].get_text(strip=True)
        time = cells[2].get_text(strip=True)
        # Some cells may be empty or not contain race information, so we check the content
        if session and date and '2023' in date:  # Adjust the condition based on actual content
            race_details = {
                'session': session,
                'date': date,
                'time': time
            }
            race_schedule.append(race_details)

# Print the extracted schedule
for race in race_schedule:
    print(race)
```

```python

def extract_f1_schedule(soup):
    # 初始化变量来保存跨行标题
    current_title = None
    schedule = []

    # 找到所有的行
    table_rows = soup.find_all('tr')
    for row in table_rows:
        # 在当前行中找到所有单元格
        cells = row.find_all('td')
        
        # 检查是否是带有标题的行（假设标题行有 'rowspan' 属性）
        if cells and 'rowspan' in cells[0].attrs:
            current_title = cells[1].get_text(strip=True)
        
        # 根据单元格数量来判断是否是详细信息的行
        if len(cells) == 5:  # 假设详细信息行有5个单元格
            # 提取文本
            session = cells[0].get_text(strip=True)
            date = cells[1].get_text(strip=True)
            time = cells[2].get_text(strip=True)
            
            # 将提取的信息添加到日程列表中
            schedule.append({
                'title': current_title,
                'session': session,
                'date': date,
                'time': time
            })

    return schedule

# 加载 HTML 并创建 BeautifulSoup 对象
file_path = r"E:\JHU\SoftwareCarpentry\Final_Project\2023 Formula 1 World Championship Race Schedule on ESPN Platforms - ESPN Press Room U.S..html"

# Open and read the HTML file
with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()
# html_content = """您的 HTML 内容"""
soup = BeautifulSoup(html_content, 'html.parser')

# 提取 F1 日程
f1_schedule = extract_f1_schedule(soup)

# 打印提取的日程
for event in f1_schedule:
    print(event)
```
